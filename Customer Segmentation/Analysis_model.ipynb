```python
# 📊 Customer Segmentation & Marketing Analytics Project
# -----------------------------------------------------
# الهدف: تحليل بيانات الزبناء لاكتشاف أنماط الشراء والإنفاق 
# واستعمالها في تقسيم العملاء (Customer Segmentation)

# 🧩 استيراد المكتبات الضرورية
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 🔹 تحميل البيانات
df = pd.read_csv("customer_segmentation.csv")

# 🔹 نظرة أولية على البيانات
print(df.shape)       # عدد الصفوف والأعمدة
print(df.columns)     # أسماء الأعمدة
df.info()             # أنواع البيانات والمعلومات العامة

# 🔹 التحقق من القيم الفارغة
print(df.isna().sum())

# 🔹 حذف الصفوف اللي فيها قيم مفقودة (NaN)
df.dropna(inplace=True)

# 🔹 ملخص إحصائي للبيانات
print(df.describe())

# 🔹 توزيع القيم في الأعمدة الفئوية (Categorical)
print(df["Education"].value_counts())
print(df["Marital_Status"].value_counts())

# 🔹 تحويل عمود التاريخ إلى نوع datetime
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], dayfirst=True)

# 🔹 إنشاء عمود جديد للعمر بناءً على سنة الميلاد
df["Age"] = 2025 - df["Year_Birth"]

# 🔹 حساب عدد الأطفال الكلي (صغار + مراهقين)
df["Total_Children"] = df["Kidhome"] + df["Teenhome"]

# 🔹 حساب مجموع الإنفاق الكلي على جميع المنتجات
spend_cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 
              'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df["Total_Spending"] = df[spend_cols].sum(axis=1)

# 🔹 حساب عدد الأيام منذ أن أصبح الزبون عميلاً (Customer Lifetime)
df["Customer_Since"] = (pd.Timestamp("today") - df["Dt_Customer"]).dt.days

# --------------------------
# 📈 التحليل الاستكشافي (EDA)
# --------------------------

# 🔹 توزيع الأعمار
sns.histplot(df["Age"], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()

# 🔹 توزيع الدخل
sns.histplot(df["Income"], bins=30, kde=True)
plt.title("Income Distribution")
plt.show()

# 🔹 توزيع الإنفاق الكلي
sns.histplot(df["Total_Spending"], bins=30, kde=True)
plt.title("Total Spending Distribution")
plt.show()

# 🔹 العلاقة بين التعليم والدخل
sns.boxenplot(x="Education", y="Income", data=df)
plt.xticks(rotation=45)
plt.title("Income by Education Level")
plt.show()

# 🔹 العلاقة بين الحالة العائلية والإنفاق
sns.boxenplot(x="Marital_Status", y="Total_Spending", data=df)
plt.xticks(rotation=45)
plt.title("Spending by Marital Status")
plt.show()

# 🔹 مصفوفة الارتباط (Correlation Matrix)
corr = df[["Income", "Age", "Recency", "Total_Spending", "NumWebPurchases", "NumStorePurchases"]]
sns.heatmap(corr.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

# 🔹 Pivot Table للدخل حسب التعليم والحالة العائلية
pivot_income = df.pivot_table(values="Income", index="Education", 
                              columns="Marital_Status", aggfunc="mean")
sns.heatmap(pivot_income, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Average Income by Education and Marital Status")
plt.show()

# 🔹 الإنفاق المتوسط حسب مستوى التعليم
group1 = df.groupby("Education")["Total_Spending"].mean().sort_values(ascending=False)
group1.plot(kind="bar", color="skyblue")
plt.title("Average Spending by Education Level")
plt.ylabel("Average Total Spending")
plt.xticks(rotation=45)
plt.show()

# --------------------------
# 🎯 الحملات التسويقية
# --------------------------

# 🔹 إنشاء عمود جديد يمثل مجموع الحملات المقبولة من العميل
df["AcceptedAny"] = df[["AcceptedCmp1", "AcceptedCmp2", 
                        "AcceptedCmp3", "AcceptedCmp4", 
                        "AcceptedCmp5", "Response"]].sum(axis=1)

print(df["AcceptedAny"].unique())  # القيم المختلفة الممكنة
```

```python
# ------------------------------------------------------------
# 🎯 Customer Segmentation using K-Means Clustering
# ------------------------------------------------------------

# ✅ تحويل القيم اللي قبلات الحملات التسويقية إلى 0 أو 1
df["AcceptedAny"] = df["AcceptedAny"].apply(lambda x: 1 if x > 0 else 0)

# 🔹 التحقق من القيم الفريدة
print(df["AcceptedAny"].unique())

# ------------------------------------------------------------
# 💍 تحليل معدل قبول الحملات حسب الحالة العائلية
# ------------------------------------------------------------
group2 = df.groupby("Marital_Status")["AcceptedAny"].mean().sort_values(ascending=False)
print(group2)

# 🔹 تمثيل بياني لمعدل القبول حسب الحالة العائلية
group2.plot(kind="bar", color="orange")
plt.title("Campaign Acceptance Rate by Marital Status")
plt.ylabel("Acceptance Rate")
plt.xticks(rotation=45)
plt.show()

# ------------------------------------------------------------
# 👵👩‍🦱 تقسيم الأعمار إلى مجموعات (Age Groups)
# ------------------------------------------------------------
bins = [18, 30, 40, 50, 60, 70, 90]
labels = ["18-29", "30-39", "40-49", "50-59", "60-69", "70+"]
df["AgeGroup"] = pd.cut(df["Age"], bins=bins, labels=labels)

print(df["AgeGroup"].head())

# ------------------------------------------------------------
# 💰 تحليل متوسط الدخل حسب الفئة العمرية
# ------------------------------------------------------------
group3 = df.groupby("AgeGroup")["Income"].mean()
print(group3)

group3.plot(kind="barh", color="green")
plt.title("Average Income by Age Group")
plt.xlabel("Average Income")
plt.show()

# ------------------------------------------------------------
# 📊 إعداد الميزات (Features) لاستخدامها في التجميع (Clustering)
# ------------------------------------------------------------
features = ["Age", "Income", "Total_Spending", 
            "NumWebPurchases", "NumStorePurchases", 
            "NumWebVisitsMonth", "Recency"]

X = df[features].copy()
print(X.head())

# ------------------------------------------------------------
# ⚖️ توحيد القيم (Standardization) باش تكون في نفس المقياس
# ------------------------------------------------------------
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(X_scaled.shape)  # التأكد من الأبعاد

# ------------------------------------------------------------
# 🧩 تحديد عدد المجموعات المثالي باستخدام Elbow Method
# ------------------------------------------------------------
from sklearn.cluster import KMeans

wcss = []
for i in range(2, 10):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# 🔹 تمثيل بياني لطريقة Elbow
plt.plot(range(2, 10), wcss, marker="o")
plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()

# ------------------------------------------------------------
# ✅ تطبيق K-Means Clustering بعدد 6 مجموعات (حسب Elbow)
# ------------------------------------------------------------
kmeans = KMeans(n_clusters=6, random_state=42)
df["Cluster"] = kmeans.fit_predict(X_scaled)

# 🔹 نظرة أولية بعد إضافة عمود التجميع
print(df[["ID", "Age", "Income", "Cluster"]].head())

# ------------------------------------------------------------
# 📈 تحليل الخصائص لكل Cluster
# ------------------------------------------------------------
cluster_summary = df.groupby("Cluster")[features].mean().round(2)
print(cluster_summary)

# 🔹 عدد العملاء في كل Cluster
print(df["Cluster"].value_counts())

# ------------------------------------------------------------
# 🎨 تقليل الأبعاد باستخدام PCA لعرض المجموعات في بعدين
# ------------------------------------------------------------
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_data = pca.fit_transform(X_scaled)

# 🔹 إضافة الأعمدة الجديدة
df["PCA1"], df["PCA2"] = pca_data[:, 0], pca_data[:, 1]

# 🔹 رسم المجموعات
sns.scatterplot(x="PCA1", y="PCA2", hue="Cluster", data=df, palette="Set1")
plt.title("Customer Segmentation (PCA Visualization)")
plt.show()

# ------------------------------------------------------------
# 📊 النتائج النهائية
# ------------------------------------------------------------
print("📊 Cluster Summary:")
print(cluster_summary)

print("✅ K-Means Clustering Completed Successfully!")
```
